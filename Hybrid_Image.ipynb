```python
# -*- coding: utf-8 -*-
"""
Created on Fri Jun 11 14:49:25 2021

"""
import cv2 
from google.colab.patches import cv2_imshow # for image display
from skimage.filters import threshold_otsu


#the input images
img1 = cv2.imread('/content/sample_data/plane.bmp')
img2 = cv2.imread('/content/sample_data/bird.bmp')


#converting to greyscale
#comment these line if want RGB
img1 = cv2.cvtColor(img1, cv2.COLOR_BGR2GRAY)
img2 = cv2.cvtColor(img2, cv2.COLOR_BGR2GRAY)
cv2_imshow(img1)
cv2_imshow(img2)

temp = cv2.GaussianBlur(img1, (9,9), 5)
#cv2_imshow(temp)
highpass_image1 = img1 - temp
cv2_imshow(highpass_image1)

lowpass_image2 = cv2.GaussianBlur(img2, (71,71), 2.5)
cv2_imshow(lowpass_image2)

hybrid_image =  (highpass_image1 + lowpass_image2)

# Displaying Hybrid image
cv2_imshow(hybrid_image)
# Save Hybrid image to see the difference
cv2.imwrite('hybrid.jpg', hybrid_image)
```
